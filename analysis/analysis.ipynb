```python
def plot_bar_chart(data_pkl, data_json, ylabel, title, filename='bar_chart.png', dpi=300):
    labels = ['Выборка_10', 'Выборка_100', 'Выборка_1000', 'Выборка_10000']
    # Определяем ширину столбцов
    width = 0.35

    # Позиции столбцов на оси X
    x = np.arange(len(labels))

    # Создаем фигуру и ось
    fig, ax = plt.subplots()

    # Строим столбцы для каждого набора данных
    bars1 = ax.bar(x - width / 2, data_pkl, width, label='pickle', color='blue', alpha=0.6)
    bars2 = ax.bar(x + width / 2, data_json, width, label='TinyDB(JSON)', color='purple', alpha=0.6)

    # Добавляем подписи для оси X и Y
    ax.set_xlabel('Выборки')
    ax.set_ylabel(ylabel)
    ax.set_title(title)

    # Настройка меток на оси X
    ax.set_xticks(x)
    ax.set_xticklabels(labels)

    # Добавляем легенду
    ax.legend()

    # Сохраняем диаграмму в формате PNG с указанным разрешением
    plt.savefig(filename, dpi=dpi)

    # Отображаем диаграмму
    plt.show()
```

```python
plot_bar_chart(training_pkl['mean_time_s'], training_json['mean_time_s'], ylabel='Среднее время (сек)', title='Сравнение среднего времени при обучении (сек)')
```

```python
plot_bar_chart(training_pkl['mean_ram_mb'], training_json['mean_ram_mb'], ylabel='Средний объем оперативной памяти (Мб)', title='Сравнение среднего среднего объема оперативной памяти при обучении (Мб)')
```
